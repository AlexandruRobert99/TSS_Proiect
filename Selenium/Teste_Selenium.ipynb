```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome(ChromeDriverManager().install())

try:

    driver.get("https://www.saucedemo.com")

    username_input = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.ID, "user-name"))
    )

    username_input.send_keys("standard_user")

    password_input = driver.find_element(By.ID, "password")
    password_input.send_keys("secret_saucee")

    login_button = driver.find_element(By.ID, "login-button")
    login_button.click()

    WebDriverWait(driver, 5).until(
        EC.url_contains("inventory.html")
    )
    print("Login successful!")

except TimeoutException:
    
    error_message = driver.find_element(By.CSS_SELECTOR, "div.error-message-container.error").text
    print("Login failed due to timeout. Error message on page: ", error_message)
    
except Exception as e:
    print("Login failed: ", e)

finally:
    driver.quit()
```

```python
import unittest
import threading
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

class SauceDemoTest(unittest.TestCase):
    def setUp(self):
        self.driver = webdriver.Chrome(ChromeDriverManager().install())

    def test_login_user1(self):
        self.login_and_checkout("standard_user", "secret_sauce")

    def test_login_user2(self):
        self.login_and_checkout("problem_user", "secret_sauce")

    def login_and_checkout(self, username, password):
        driver = self.driver
        driver.get("https://www.saucedemo.com")
        driver.find_element(By.ID, "user-name").send_keys(username)
        driver.find_element(By.ID, "password").send_keys(password)
        driver.find_element(By.ID, "login-button").click()
        
        driver.get("https://www.saucedemo.com/inventory.html")
        
        items = driver.find_elements(By.CLASS_NAME, "inventory_item")
        
        for item in items:

            item_name = item.find_element(By.CLASS_NAME, "inventory_item_name").text
        
            add_to_cart_button = item.find_element(By.CLASS_NAME, "btn_inventory")
        
            add_to_cart_button.click()
        
            print(f"Added {item_name} to {username}'s cart.")

        
        shopping_cart_button = driver.find_element(By.CLASS_NAME, "shopping_cart_link")
        shopping_cart_button.click()

        items_counter = driver.find_element(By.CLASS_NAME, "shopping_cart_badge").text
        print(f"User {username} has {items_counter} items in the cart.")
        
        checkout_button = driver.find_element(By.ID, "checkout")
        checkout_button.click()
        
        current_url = driver.current_url
        print("URL after checkout:", current_url)
        
        
        WebDriverWait(driver, 10).until(EC.url_contains("checkout-step-one.html"))
        

        driver.find_element(By.ID, "first-name").send_keys("John")
        
        driver.find_element(By.ID, "last-name").send_keys("Doe")
        
        driver.find_element(By.ID, "postal-code").send_keys("12345")

        driver.find_element(By.ID, "continue").click()
        
        WebDriverWait(driver, 10).until(EC.url_contains("checkout-step-two.html"))
        
        final_url = driver.current_url
        print(f"URL after checkout step two of {username}:", final_url)
        
        finish_button = driver.find_element(By.ID, "finish")
        finish_button.click()
        
        WebDriverWait(driver, 10).until(EC.url_contains("checkout-complete.html"))
        print(f"Checkout complete for user {username}. URL:", driver.current_url)

        print(f"Completed test for {username}")

    def tearDown(self):
        self.driver.quit()

def run_test(test_name):
    suite = unittest.TestSuite()
    suite.addTest(SauceDemoTest(test_name))
    runner = unittest.TextTestRunner()
    runner.run(suite)

if __name__ == "__main__":
    
    user1_thread = threading.Thread(target=run_test, args=("test_login_user1",))
    user2_thread = threading.Thread(target=run_test, args=("test_login_user2",))

    user1_thread.start()
    user2_thread.start()

    user1_thread.join()
    user2_thread.join()
```
